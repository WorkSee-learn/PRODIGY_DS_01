```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/JUPYTER/Data Files/dict-age-groups.csv')
```

```python
print(df)
```

```python
total_males = df['male'].sum()
total_females = df['female'].sum()
```

```python
print(total_males)
print(total_females)
```

```python
gender_distribution = pd.DataFrame({
    'gender': ['Male', 'Female'],
    'count': [total_males, total_females]
})
```

```python
print(gender_distribution)
```

```python
plt.bar(gender_distribution['gender'], gender_distribution['count'], color=['blue', 'pink'], alpha=0.7)
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
```

```python
# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bars for males and females
bar_male = ax.bar(df['id'], df['male'], label='Male', color='blue', alpha=0.7)
bar_female = ax.bar(df['id'], df['female'], bottom=df['male'], label='Female', color='pink', alpha=0.7)

# Adding titles and labels
ax.set_xlabel('Age Group')
ax.set_ylabel('Population')
ax.set_title('Population Distribution by Gender and Age Group (Stacked Bar Chart)')
ax.legend()

plt.show()
```

```python
# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bar positions for grouped bar chart
bar_width = 0.35
bar_positions = np.arange(len(df['id']))

# Bars for males and females
bar_male = ax.bar(bar_positions - bar_width/2, df['male'], bar_width, label='Male', color='red', alpha=0.7)
bar_female = ax.bar(bar_positions + bar_width/2, df['female'], bar_width, label='Female', color='green', alpha=0.7)

# Adding titles and labels
ax.set_xlabel('Age Group')
ax.set_ylabel('Population')
ax.set_title('Population Distribution by Gender and Age Group (Grouped Bar Chart)')
ax.set_xticks(bar_positions)
ax.set_xticklabels(df['id'])
ax.legend()

plt.show()
```
